```python
def is_na(row):
    return pd.isnull(row['NIK']) or pd.isnull(row['Tgl.Lahir']) or pd.isnull(row['Jenis Kelamin'])


def valid_zs(row):
    nik_str = str(row['NIK'])

    # NIK harus 16 digit dan semua karakter angka
    if len(nik_str) != 16 or not nik_str.isdigit():
        return "Invalid"


    # Ekstraksi tanggal, bulan, tahun dari NIK
    tanggal_nik = nik_str[6:8]
    bulan_nik = nik_str[8:10]
    tahun_nik = nik_str[10:12]


    try:
        tanggal_lahir = row['Tgl.Lahir'].strftime('%d')
        bulan_lahir = row['Tgl.Lahir'].strftime('%m')
        tahun_lahir = row['Tgl.Lahir'].strftime('%y')  # dua digit terakhir
    except:
        return "NA"


    try:
        tgl_nik_int = int(tanggal_nik)
        tgl_lahir_int = int(tanggal_lahir)
    except:
        return "NA"


    # Validasi bulan dan tahun
    month_year_valid = (tahun_nik == tahun_lahir) and (bulan_nik == bulan_lahir)


    # Validasi tanggal sesuai gender
    if row['Jenis Kelamin'] == "Laki-laki":
        day_valid = tgl_nik_int == tgl_lahir_int
    elif row['Jenis Kelamin'] == "Perempuan":
        day_valid = tgl_nik_int == (tgl_lahir_int + 40)
    else:
        day_valid = False


    return "Valid NIK" if month_year_valid and day_valid else "Invalid"


# Apply
data['nik_date_status'] = data.apply(lambda row: "NA" if is_na(row) else valid_zs(row), axis=1)



# Apply NA first
data['nik_date_status'] = data.apply(lambda row: "NA" if is_na(row) else valid_zs(row), axis=1)

# Group by puskesmas_names untuk summary per puskesmas
grouped = data.groupby('puskesmas_names')

# List untuk menyimpan summary per puskesmas
summaries = []

for puskesmas, group in grouped:
    counts = group['nik_date_status'].value_counts()
    percentages = group['nik_date_status'].value_counts(normalize=True) * 100
    summary = pd.DataFrame({
        "Puskesmas": puskesmas,
        "Status": counts.index,
        "Count": counts.values,
        "Percentage": [f"{x:.2f}%" for x in percentages]
    })
    summaries.append(summary)

# Gabungkan semua summary menjadi satu DataFrame
all_summary = pd.concat(summaries, ignore_index=True)

# Print hasil dalam format markdown
print(all_summary.to_markdown(index=False))
```

```python
def plot_hist_per_nakes_table(subset, var, group_by):
    # Group the data to get the total patients for each group
    group_count = subset.groupby(group_by).size().reset_index(name='Total_Pasien')

    # Group by the variable and group_by column to get frequencies
    freq = subset.groupby([var, group_by]).size().reset_index(name='Jumlah')

    # Merge the frequency data with the total patient count per group
    freq = freq.merge(group_count, on=group_by)

    # Calculate the percentage per group
    freq['Persentase Perawat / Bidan / Nutrisionist / Sanitarian'] = ((freq['Jumlah'] / freq['Total_Pasien']) * 100).round(2)

    # Sort the data by 'Jumlah' and get the top 5
    top_5 = freq.sort_values(by='Jumlah', ascending=False).head(5)

    return top_5[[var, group_by, 'Jumlah', 'Persentase Perawat / Bidan / Nutrisionist / Sanitarian']]

# Dapatkan daftar puskesmas unik
unique_puskesmas = data['puskesmas_names'].unique()

# List untuk menyimpan hasil per puskesmas
results = []

for puskesmas in unique_puskesmas:
    # Filter data untuk puskesmas ini
    subset = data[data['puskesmas_names'] == puskesmas]

    # Jalankan fungsi pada subset
    top_5_subset = plot_hist_per_nakes_table(subset, 'Tinggi', 'Perawat / Bidan / Nutrisionist / Sanitarian')

    # Tambahkan kolom Puskesmas
    top_5_subset['Puskesmas'] = puskesmas

    # Reorder kolom jika diperlukan
    top_5_subset = top_5_subset[['Puskesmas', 'Tinggi', 'Perawat / Bidan / Nutrisionist / Sanitarian', 'Jumlah', 'Persentase Perawat / Bidan / Nutrisionist / Sanitarian']]

    results.append(top_5_subset)

# Gabungkan semua hasil menjadi satu DataFrame
top_5_tinggi_dokter = pd.concat(results, ignore_index=True)

# Tampilkan top 10 (atau seluruhnya jika kurang dari 10)
print(top_5_tinggi_dokter.head(30).to_markdown(index=False))
```

```python
import os
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
from openpyxl import load_workbook
from openpyxl.styles import Font, PatternFill, Alignment
from openpyxl.utils import get_column_letter

# ============================================================
# 1 SETUP OUTPUT & STYLE
# ============================================================

save_path = os.path.join('/content', 'E-PUSKESMAS_Desember_15-21_all_reports.xlsx')

header_font = Font(bold=True, size=14)
subheader_font = Font(bold=True, size=11)
header_fill = PatternFill(start_color="90EE90", end_color="90EE90", fill_type="solid")
title_alignment = Alignment(wrap_text=True, vertical='center')
header_alignment = Alignment(wrap_text=True, vertical='center')

numeric_vars = [
    'Tinggi', 'Berat Badan', 'Lingkar Perut',
    'Sistole', 'Diastole', 'Nafas', 'Detak Nadi', 'Suhu'
]
hcw_col = 'Perawat / Bidan / Nutrisionist / Sanitarian'

# Normalize names
data['puskesmas_names'] = data['puskesmas_names'].astype(str).str.upper().str.strip()

# ============================================================
# 1.1  FILTER PUSKESMAS YANG AKAN DI-GENERATE
# ============================================================

allowed_puskesmas = [
    "PUSKESMAS SERAYU LARANGAN",
    "PUSKESMAS BOBOTSARI",
    "PUSKESMAS KARANGJAMBU",
    "PUSKESMAS KARANG TENGAH",
    "PUSKESMAS BOJONG",
    "PUSKESMAS KALIGONDANG",
    "PUSKESMAS BOJONGSARI",
    "PUSKESMAS PURBALINGGA",
    "PUSKESMAS KARANGANYAR",
    "PUSKESMAS PADAMARA",
    "PUSKESMAS REMBANG",
    "PUSKESMAS PENGADEGAN"
]

puskesmas_list = [p for p in allowed_puskesmas if p in data['puskesmas_names'].unique()]

# ============================================================
# 2 VALIDATION FUNCTIONS
# ============================================================

def is_na(row):
    return pd.isnull(row['NIK']) or pd.isnull(row['Tgl.Lahir']) or pd.isnull(row['Jenis Kelamin'])

def valid_zs(row):
    nik_str = str(row['NIK']).strip()
    if len(nik_str) != 16 or not nik_str.isdigit():
        return "Invalid"

    tanggal_nik = nik_str[6:8]
    bulan_nik = nik_str[8:10]
    tahun_nik = nik_str[10:12]

    try:
        tanggal_lahir = row['Tgl.Lahir'].strftime('%d')
        bulan_lahir = row['Tgl.Lahir'].strftime('%m')
        tahun_lahir = row['Tgl.Lahir'].strftime('%y')
    except:
        return "NA"

    try:
        tgl_nik_int = int(tanggal_nik)
        tgl_lahir_int = int(tanggal_lahir)
    except:
        return "NA"

    month_year_valid = (tahun_nik == tahun_lahir) and (bulan_nik == bulan_lahir)

    if row['Jenis Kelamin'] == "Laki-laki":
        day_valid = tgl_nik_int == tgl_lahir_int
    elif row['Jenis Kelamin'] == "Perempuan":
        day_valid = tgl_nik_int == (tgl_lahir_int + 40)
    else:
        day_valid = False

    return "Valid NIK" if month_year_valid and day_valid else "Invalid"

# ============================================================
# 3 LOOP PER PUSKESMAS (ONLY FILTERED ONES)
# ============================================================

with pd.ExcelWriter(save_path, engine='openpyxl') as writer:
    writer.book.create_sheet("TEMP_INIT")

    for pusk in tqdm(puskesmas_list, desc="Processing Puskesmas"):

        sub = data[data['puskesmas_names'] == pusk].copy()
        n = len(sub)

        # --- Character Count Validation ---
        sub['NIK_str'] = sub['NIK'].astype(str).str.strip()
        sub['nik_char_status'] = sub['NIK_str'].apply(
            lambda x: 'NA' if x.lower() in ['nan', 'none', '<na>']
            else ('Valid' if len(x) == 16 else 'Invalid')
        )
        vc_char = sub['nik_char_status'].value_counts(dropna=False)
        val_char = pd.DataFrame({
            'Status': vc_char.index,
            'Count': vc_char.values,
            'Percentage': (vc_char / vc_char.sum() * 100).round(2)
        })

        # --- Date Match Validation ---
        sub['nik_date_status'] = sub.apply(lambda row: "NA" if is_na(row) else valid_zs(row), axis=1)
        vc_date = sub['nik_date_status'].value_counts(dropna=False)
        val_date = pd.DataFrame({
            'Status': vc_date.index,
            'Count': vc_date.values,
            'Percentage': (vc_date / vc_date.sum() * 100).round(2)
        })

        # --- Phone Validation ---
        if 'No Telp' in sub.columns:
            telp = sub['No Telp'].astype(str).str.strip()
            telp = telp.where(~telp.str.lower().isin(['nan', 'nat', 'none', '<na>']), np.nan)
            telp_status = np.where(
                telp.isna(), 'NA',
                np.where(telp.str.len() >= 10, 'Valid', 'Invalid')
            )
            vc_phone = pd.Series(telp_status).value_counts()
            val_phone = pd.DataFrame({
                'Status': vc_phone.index,
                'Count': vc_phone.values,
                'Percentage': (vc_phone / vc_phone.sum() * 100).round(2)
            })
        else:
            val_phone = pd.DataFrame(columns=['Status','Count','Percentage'])

        # --- Diagnosis Count ---
        if {'ICD-X 1','Diagnosa 1'}.issubset(sub.columns):
            diagnosis_counts = sub.groupby(['ICD-X 1','Diagnosa 1']).size().reset_index(name='Jumlah Pasien')
        else:
            diagnosis_counts = pd.DataFrame(columns=['ICD-X 1','Diagnosa 1','Jumlah Pasien'])

        # --- Nakes ---
        if 'Dokter / Tenaga Medis' in sub.columns:
            nakes = (
                sub['Dokter / Tenaga Medis']
                .value_counts().rename_axis('Dokter / Tenaga Medis')
                .reset_index(name='Jumlah Pasien')
            )
        else:
            nakes = pd.DataFrame(columns=['Dokter / Tenaga Medis','Jumlah Pasien'])

        # --- Completeness Calculation ---
        avg_cols = [
            'NIK','No Telp','Jenis Kelamin',
            'Dokter / Tenaga Medis', 'Perawat / Bidan / Nutrisionist / Sanitarian',
            'Tinggi','Berat Badan','Lingkar Perut','Sistole','Diastole',
            'Nafas','Detak Nadi','Suhu','Diagnosa 1'
        ]
        reg_cols = ['NIK', 'No Telp', 'Tgl.Lahir']
        hcw_cols = [
            'Dokter / Tenaga Medis','Perawat / Bidan / Nutrisionist / Sanitarian',
            'Diagnosa 1','Tinggi','Berat Badan','Lingkar Perut',
            'Sistole','Diastole','Nafas','Detak Nadi','Suhu'
        ]

        def completeness_rate(df, cols):
            cols = [c for c in cols if c in df.columns]
            if not cols:
                return 0.0
            total_cells = len(df) * len(cols)
            filled_cells = df[cols].notna().sum().sum()
            return round(filled_cells / total_cells * 100, 2)

        avg_comp = completeness_rate(sub, avg_cols)
        reg_comp = completeness_rate(sub, reg_cols)
        hcw_comp = completeness_rate(sub, hcw_cols)

        val_comp = pd.DataFrame({
            'Category': ['Average Completeness (%)','Registration Completeness (%)','HCWs Missing Variables (%)'],
            'Value': [avg_comp, reg_comp, hcw_comp]
        })

        # --- Accuracy Tables ---
        def acc_all(subb, var):
            if var not in subb.columns:
                return pd.DataFrame(columns=[var, 'Jumlah', 'Percentage'])
            valid = subb.dropna(subset=[var])
            if valid.empty:
                return pd.DataFrame(columns=[var, 'Jumlah', 'Percentage'])
            freq = valid[var].value_counts().reset_index()
            freq.columns = [var, 'Jumlah']
            freq['Percentage'] = (freq['Jumlah'] / freq['Jumlah'].sum() * 100).round(2)
            return freq

        acc_all_dict = {v: acc_all(sub, v) for v in numeric_vars}

        # ======================================================
        # 4 WRITE TO EXCEL
        # ======================================================

        tables = {
            "Number of Patient": n,
            "Number of Patients / Diagnosis": diagnosis_counts,
            "Number of Patients per Healthcare Worker (Nakes)": nakes,
            "Completeness Summary": val_comp,
            "Validation NIK (Character Count)": val_char,
            "Validation NIK (Date Match)": val_date,
            "Phone Number Validation": val_phone,
        }
        for var in numeric_vars:
            tables[f"Accuracy ALL ({var})"] = acc_all_dict[var]

        sheet_name = pusk[:31]
        if sheet_name not in writer.book.sheetnames:
            writer.book.create_sheet(sheet_name)
        ws = writer.book[sheet_name]

        start_row = 1

        for title, df in tables.items():
            dfw = df if isinstance(df, pd.DataFrame) else pd.DataFrame({"Jumlah Pasien": [df]})

            tcell = ws.cell(row=start_row, column=1)
            tcell.value = title
            tcell.font = header_font
            tcell.fill = header_fill
            tcell.alignment = title_alignment

            dfw.to_excel(writer, sheet_name=sheet_name, startrow=start_row, index=False, header=True)

            if not dfw.empty:
                end_col = len(dfw.columns)
                for cidx in range(1, end_col + 1):
                    hcell = ws.cell(row=start_row + 1, column=cidx)
                    hcell.font = subheader_font
                    hcell.fill = header_fill
                    hcell.alignment = header_alignment

                start_row += 1 + 1 + len(dfw) + 4
            else:
                start_row += 2 + 4

        # Auto Column Width
        for column in ws.columns:
            max_len = max((len(str(cell.value)) for cell in column if cell.value), default=0)
            ws.column_dimensions[get_column_letter(column[0].column)].width = min(max_len + 2, 50)

    # Remove temp sheet
    if "TEMP_INIT" in writer.book.sheetnames:
        writer.book.remove(writer.book["TEMP_INIT"])

wb = load_workbook(save_path)
wb.save(save_path)

print(f" File laporan tersimpan di: {save_path}")
```

```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import Font
from openpyxl.utils import get_column_letter
from datetime import datetime, date

# --- Helper untuk memastikan DataFrame valid ---
def ensure_df(x, title=None):
    if isinstance(x, pd.DataFrame):
        return x
    elif isinstance(x, pd.Series):
        return x.to_frame()
    else:
        return pd.DataFrame({title if title else "Value": [x]})

# --- Daftar tabel yang akan digabung ---
tables = {
    "Table of Valid NIK based on Character Count": valid_nik,
    "Table of Invalid NIK based on Character Count": invalid_nik,
    "Table of NA NIK based on Character Count": NA_nik,
    "NIK Date Match Verification": summary,
    "Table of Valid NIK Dates": valid_nik_dates,
    "Table of Invalid NIK Dates": invalid_nik_dates,
    "Table of NA NIK Dates": na_nik_dates,
    "Table of Valid Phone Numbers": valid_telp,
    "Table of Invalid Phone Numbers": invalid_telp,
    "Table of NA Phone Numbers": na_telp,
}

save_path = "Validation_Desember_15-21_PUSKESMAS.xlsx"

# --- Tulis semua tabel ke Excel ---
with pd.ExcelWriter(save_path, engine="openpyxl") as writer:
    start_row = 0
    for title, table in tables.items():
        df = ensure_df(table, title).copy()

        # Urutkan berdasarkan nama puskesmas (jika ada)
        if "puskesmas_names" in df.columns:
            df = df.sort_values(by="puskesmas_names", ascending=True).reset_index(drop=True)

        # Format kolom NIK sebagai teks
        if "NIK" in df.columns:
            df["NIK"] = df["NIK"].astype(str).apply(lambda x: f"\u00A0{x}")

        # Konversi kolom Tanggal jadi DATE (bukan datetime)
        if "Tanggal" in df.columns:
            df["Tanggal"] = pd.to_datetime(df["Tanggal"], errors="coerce").dt.date

        # Pastikan Sheet1 ada
        if "Sheet1" not in writer.book.sheetnames:
            writer.book.create_sheet("Sheet1")
        worksheet = writer.sheets.get("Sheet1", writer.book["Sheet1"])

        # Tulis judul tabel
        worksheet.cell(row=start_row + 1, column=1, value=title)
        start_row += 1

        # Tulis DataFrame
        df.to_excel(writer, sheet_name="Sheet1", startrow=start_row, index=False)
        start_row += len(df) + 5  # jarak antar tabel

# --- Buka kembali file untuk format judul & kolom Tanggal ---
wb = load_workbook(save_path)
ws = wb["Sheet1"]

# Format judul tabel besar & tebal
row = 1
for title, table in tables.items():
    ws.cell(row=row, column=1).font = Font(size=18, bold=True)
    row += 1
    df = ensure_df(table, title)
    row += len(df) + 5

# Format kolom 'Tanggal' agar tampil dd/mm/yyyy TANPA waktu
date_style = "dd/mm/yyyy"

for col in ws.iter_cols():
    if col[0].value == "Tanggal":  # cari kolom Tanggal dari header
        col_idx = col[0].column
        for cell in ws[get_column_letter(col_idx)][1:]:  # skip header
            # Pastikan nilainya adalah DATE, bukan datetime
            if isinstance(cell.value, datetime):
                cell.value = cell.value.date()
                cell.number_format = date_style
            elif isinstance(cell.value, str):
                try:
                    val = pd.to_datetime(cell.value).date()
                    cell.value = val
                    cell.number_format = date_style
                except Exception:
                    pass
            elif isinstance(cell.value, date):
                cell.number_format = date_style

# Simpan hasil akhir
wb.save(save_path)
print(f" File laporan berhasil disimpan di: {save_path}")
```
